```python
#Bar plot in Seaborn library
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = df1)
plt.show()
```

```python
#IN bargraph categorical data
```

```python
df1
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex")
plt.show()
```

```python
#order parameter
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex",order = order1)
plt.show()
```

```python
#hue_order
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex",order = order1, hue_order = ['Female', 'Male'])
plt.show()
```

```python
order1 = ['Dream', 'Tongersen', 'Boscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, color = 'red')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = 'sex', palette = 'Greys')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = 'sex', palette = 'prism')
plt.show()
```

```python
#saturation #fade karna colour ko 0 to 1
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = 'sex', palette = 'prism', saturation = 0.5)
plt.show()
```
