```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#using matplotlib

var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]

plt.plot(var, var1)
plt.show()
```

```python
#using seaborn

var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]

sns.lineplot(x = var, y = var1)
plt.show()
```

```python
#using pandas

import pandas as pd

df = pd.DataFrame({'var' : var, 'var1' : var1})
sns.lineplot(x = 'var', y = 'var1', data = df)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = sns.load_dataset("penguins")
df1
```

```python
df1['island']
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm, data = df1)
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = 'sex')
plt.show()
```

```python
#order

order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex",order = order1)
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex",order = order1, hue_order = ['Female', 'Male'])
plt.show()
```

```python
order1 = ['Dream', 'Tongersen', 'Boscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, color = 'red')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = 'sex', palette = 'Greys')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = 'sex', palette = 'prism', saturation = 0.5)
plt.show()
```
