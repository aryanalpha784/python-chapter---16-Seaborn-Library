# Bar Plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("penguins")
df
```

```python
df['island']
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = df)
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = df, hue = 'sex')
plt.show()
```

```python
order = ['Dream', 'Torgersen', 'Biscoe island']
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex", order = order, hue_order = ['Female', 'Male'], palette = 'grey', saturation = 0.5)
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm', data = df, color = 'red')
plt.show()
```

```python
sns.displot(df['flipper_length_mm'])
```

```python
sns.displot(df['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240], color = 'red', kde = True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ADANIPORTS.csv", parse_dates = True)
```

```python
df
```

```python
df.head()
```

```python
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()
```

```python
ax = plt.axes(projection = '3d')
ax.scatter(df.index, df['H-L'], df['100MA'])
ax.set_xlabel("Index")
ax.set_ylabel("H-L")
ax.set_zlabel("100MA")

sns.set_style("darkgrid")
plt.show()
```

```python
import numpy as np

z1 = np.linspace(0,10,100)
x1 = np.cos(4 * z1)
y1 = np.sin(4 * z1)

ax = plt.axes(projection = '3d')
ax.plot(x1, y1, z1)

sns.set_style("whitegrid")
plt.show()
```

```python
def return_z(x, y):
    return 58 - (x**2 + y**2)
```

```python
x1, y1 = np.linspace(-5, 5, 50), np.linspace(-5, 5, 50)

x1, y1 = np.meshgrid(x1, y1)
z1 = return_z(x1, y1)

ax = plt.axes(projection = '3d')
ax.plot_surface(x1, y1, z1)


sns.set_style('whitegrid')
plt.show()
```
